```python
from scipy.io.wavfile import read, write
import numpy as np
import IPython

sr, audio = read('/content/drive/MyDrive/klokjev2/gm_bitch.wav')   #we used this file to test the cut silence
 
def cut_silence(signal=audio, sr=sr, threshold=0.1, output='output.wav'):
    max_ampl = max(signal)   #find the max amplitude within the signal array (both + and -)
    min_ampl = min(signal)
    abs_max = max(abs(max_ampl), abs(min_ampl))  # we need the *absolute* (aka no minus sign) maximum to calculate the threshold
    cutoff = abs_max*threshold  #anything below the threshold will be counted as silence

    above = np.where(signal > cutoff)   #find first & last instance of signal that's above threshold
    below = np.where(signal < -cutoff)  #find first & last instance of signal that's below -threshold
    min1 = min(above[0][0],below[0][0])  #which of the two is smaller: that's the instance we need for the beginning
    max1 = max(above[0][-1], below[0][-1]) #which last sample is larger: that's the instance we need for the beginning
    new_sig = signal[min1: max1]
    write(output, sr, new_sig)
    
```
